##Reading the data from text file

```
with open('text.txt') as file:
    text = file.read()

print(text)
```

```
characters = list(set(text))
vocab_length = len(characters)
print(characters)
print(f'Number of unique characters: {vocab_length}')
```

```
char_to_int = dict([(char, i) for i, char in enumerate(characters)])
char_to_int
```

```
int_to_char = dict([(i, char) for i, char in enumerate(characters)])
int_to_char
```

```
def toint(sentence):
    return [char_to_int[char] for char in sentence]
```

```
def tosentence(ints):
    return ''.join([int_to_char[_int] for _int in ints])
```

```
int_text = toint(text)
int_text
```

```
print(f'Total sequence length of the text: {len(int_text)}')
print(f'First 20 integers: {int_text[:20]}')
print(f'First 20 characters: {tosentence(int_text[:20])}')
```

```
def one_hot(ints, vocab_length = vocab_length):
    outputs = nd.zeros((len(ints), vocab_length))
    for i, idx in enumerate(ints):
        outputs[i, idx] = 1.0
    return outputs
```

```
print('Before one hot:')
print(int_text[:20])
print('After one hot:')
print(one_hot(int_text[:20]))
```

```
def onehot_to_text(embeddings):
    outputs = ""
    ints = nd.argmax(embeddings, axis = 1).asnumpy()
    for i in ints:
        outputs += int_to_char[i]
    return outputs
```

```
onehot = one_hot(int_text[:20])
print('First 20 characters in Onehot:')
print(onehot)
print('\nAfter converting First 20 one hot embeddings to text:')
print(f'"{onehot_to_text(onehot)}"')
```

##Defining Encoder Parameters

```
num_inputs = vocab_length
num_hidden = 512
num_outputs = vocab_length

eWxz = nd.random.normal(shape = (num_inputs, num_hidden), ctx = ctx) * .001
eWxr = nd.random.normal(shape = (num_inputs, num_hidden), ctx = ctx) * .001
eWxh = nd.random.normal(shape = (num_inputs, num_hidden), ctx = ctx) * .001

eWhz = nd.random.normal(shape = (num_hidden, num_hidden), ctx = ctx) * .001
eWhr = nd.random.normal(shape = (num_hidden, num_hidden), ctx = ctx) * .001
eWhh = nd.random.normal(shape = (num_hidden, num_hidden), ctx = ctx) * .001

ebz = nd.random.normal(shape = num_hidden, ctx = ctx) * .001
ebr = nd.random.normal(shape = num_hidden, ctx = ctx) * .001
ebh = nd.random.normal(shape = num_hidden, ctx = ctx) * .001

eWhy = nd.random.normal(shape = (num_hidden,num_outputs), ctx = ctx) * .001
eby = nd.random.normal(shape = num_outputs, ctx = ctx) * .001

encoder_params = [eWxz, eWxr, eWxh, eWhz, eWhr, eWhh, ebz, ebr, ebh, eWhy, eby]
```

##Defining Encoder

```
def softmax(linear):
    exp = nd.exp((linear - nd.max(linear, axis = 1).reshape((-1, 1))))
    partition = nd.sum(exp, axis = 1).reshape((-1, 1))
    return exp / partition
```

```
def encode(inputs, h):
    outputs = []
    for X in inputs:
        z = nd.sigmoid(nd.dot(X, eWxz) + nd.dot(h, eWhz) + ebz)
        r = nd.sigmoid(nd.dot(X, eWxr) + nd.dot(h, eWhr) + ebr)
        
        g = nd.tanh(nd.dot(X, eWxh) + nd.dot(r * h, eWhh) + ebh)
        h = z * h + (1 - z) * g

        linear = nd.dot(h, eWhy) + eby
        output = softmax(linear)
        outputs.append(output)
    return outputs, h
```

```
print(train_features[0].shape)
h = nd.zeros((1, num_hidden)).as_in_context(ctx)
encoded, h = encode(train_features[0].as_in_context(ctx), h)
print(len(encoded), encoded[0].shape)
print(h.shape)
```

##Defining Decoder Parameters

```
num_inputs = vocab_length
num_hidden = 512
num_outputs = vocab_length

dWxz = nd.random.normal(shape = (num_inputs, num_hidden), ctx = ctx) * .001
dWxr = nd.random.normal(shape = (num_inputs, num_hidden), ctx = ctx) * .001
dWxh = nd.random.normal(shape = (num_inputs, num_hidden), ctx = ctx) * .001

dWhz = nd.random.normal(shape = (num_hidden, num_hidden), ctx = ctx) * .001
dWhr = nd.random.normal(shape = (num_hidden, num_hidden), ctx = ctx) * .001
dWhh = nd.random.normal(shape = (num_hidden, num_hidden), ctx = ctx) * .001

dbz = nd.random.normal(shape = num_hidden, ctx = ctx) * .001
dbr = nd.random.normal(shape = num_hidden, ctx = ctx) * .001
dbh = nd.random.normal(shape = num_hidden, ctx = ctx) * .001

dWhy = nd.random.normal(shape = (num_hidden,num_outputs), ctx = ctx) * .001
dby = nd.random.normal(shape = num_outputs, ctx = ctx) * .001

decoder_params = [dWxz, dWxr, dWxh, dWhz, dWhr, dWhh, dbz, dbr, dbh, dWhy, dby]
```

## Defining Decoder

```
def decode(inputs, h):
    outputs = []
    X = inputs[-1]
    for i in range(len(inputs)):
        z = nd.sigmoid(nd.dot(X, dWxz) + nd.dot(h, dWhz) + dbz)
        r = nd.sigmoid(nd.dot(X, dWxr) + nd.dot(h, dWhr) + dbr)
        
        g = nd.tanh(nd.dot(X, dWxh) + nd.dot(r * h, dWhh) + dbh)
        h = z * h + (1 - z) * g

        linear = nd.dot(h, dWhy) + dby
        X = softmax(linear)
        outputs.append(X)
        X = X.argmax(1).one_hot(vocab_length)
    return outputs, h
```

```
decoded, h = decode(encoded, h)
print(len(decoded), decoded[0].shape)
print(h.shape)
```

```
def seq_to_seq(inputs, h):
    encoded, h = encode(inputs, h)
    decoded, _ = decode(encoded, h)
    return decoded
```

##Defining Generator

```
def generate(prefix):
    string = ''

    int_prefix = [char_to_int[char] for char in prefix]
    input_sequence = one_hot(int_prefix).as_in_context(ctx)

    h = nd.zeros(shape = (1, num_hidden), ctx = ctx)
    
    decoded = seq_to_seq(input_sequence, h)
    
    for d in decoded:
        choice = int(d[0].argmax(0).asscalar())
        string += int_to_char[choice]
    return string
```

```
print(generate("Artificial intelligence"))
```

```
def cross_entropy(pred, label):
    return -nd.sum(label * nd.log(pred))
```

```
def objective(outputs, labels):
    total_loss = 0.
    for (output, label) in zip(outputs,labels):
        total_loss = total_loss + cross_entropy(output, label)
    return total_loss / len(outputs)
```

```
params = encoder_params + decoder_params
vs = []
sqrs = []

for param in params:
    param.attach_grad()
    vs.append(param.zeros_like())
    sqrs.append(param.zeros_like())
    
def adam(t, params, vs = vs, sqrs = sqrs, lr = 0.001):
    beta1 = 0.9
    beta2 = 0.999
    eps = 1e-8

    for param, v, sqr in zip(params, vs, sqrs):
        grad = param.grad / batch_size

        v[:] = beta1 * v + (1. - beta1) * grad
        sqr[:] = beta2 * sqr + (1. - beta2) * nd.square(grad + 1e-8)

        v_bias_corr = v / (1. - beta1 ** t)
        sqr_bias_corr = sqr / (1. - beta2 ** t)

        update = lr * v_bias_corr / (nd.sqrt(sqr_bias_corr + 1e-8) + eps)
        param[:] = param - update
```

```
def sgd(params, lr):
    for param in params:
        param[:] = param - lr * param.grad
```

```
epochs = 200
learning_rate = .001
losses = []

t = 0

for epoch in range(epochs):
    if (epoch + 1) % 50 == 0:
        learning_rate = learning_rate / 1.5
    state = nd.zeros(shape = (batch_size, num_hidden), ctx = ctx)
    for i in range(batches):
        t += 1
        features = train_features[i].as_in_context(ctx)
        labels = train_labels[i].as_in_context(ctx)
        with autograd.record():
            outputs = seq_to_seq(features, state)
            loss = objective(outputs, labels)
            loss.backward()
        adam(t, params, lr = learning_rate)
    losses.append(loss.asscalar())
    print(f'\nEpoch: {epoch} | Loss: {loss.asscalar():.5f}')
    print(generate("Artificial intelligence"))
    print('-'*50)
    print(generate("Research in AI has focused chiefly on"))
```
