```
import torch
import matplotlib.pyplot as plt
```

## __Defining ArcTan function__

```
xs = torch.arange(-10, 10, 0.1)
ys = torch.atan(xs)
```

```
plt.plot(xs, ys)
plt.xlabel('Values of x')
plt.ylabel('Values of ArcTan of x')
plt.title('ArcTan Function')
plt.show()
```

## __Now defining derivative of arctan funciton__

```
def arctan_derivative(x):
    return 1.0/((x**2) + 1.0)
```

```
dxs = torch.arange(-10, 10, 0.1)
dys = arctan_derivative(xs)
```

```
plt.plot(dxs, dys)
plt.xlabel('Values of x')
plt.ylabel('Values of derivatives of ArcTan of x')
plt.title('ArcTan Derivative')
plt.show()
```

## __Plotting Arctan function and it's derivative together__

```
plt.figure(figsize = (10, 5))
plt.plot(xs, ys, label = 'ArcTan function')
plt.plot(dxs, dys, label = 'ArcTan function derivative')
plt.xlabel('Values of x')
plt.ylabel('Values of ArcTan of x and its derivatives')
plt.title("ArcTan Function and it's derivative")
plt.legend()
plt.show()
```
