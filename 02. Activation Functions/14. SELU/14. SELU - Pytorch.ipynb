```
import torch
import matplotlib.pyplot as plt
```

## __Defining Scaled Exponential Linear Unit function__

```
global lamda, alpha
lamda = 1.0507
alpha = 1.67326

def selu(x):
    return lamda * torch.where(x < 0, alpha * (torch.exp(x) - 1.0), x)
```

```
xs = torch.arange(-10, 10, 0.1)
ys = selu(xs)
```

```
plt.plot(xs, ys)
plt.xlabel('Values of x')
plt.ylabel('Values of SELU of x')
plt.title('SELU Function')
plt.show()
```

## __Now defining derivative of selu funciton__

```
def selu_derivative(x):
    return lamda * torch.where(x < 0, alpha * torch.exp(x), torch.ones_like(x))
```

```
dxs = torch.arange(-10, 10, 0.1)
dys = selu_derivative(dxs)
```

```
plt.plot(dxs, dys)
plt.xlabel('Values of x')
plt.ylabel('Values of derivatives of SELU of x')
plt.title('SELU Derivative')
plt.show()
```

## __Plotting SELU function and it's derivative together__

```
plt.figure(figsize = (10, 5))
plt.plot(xs, ys, label = 'SELU function')
plt.plot(dxs, dys, label = 'SELU function derivative')
plt.xlabel('Values of x')
plt.ylabel('Values of SELU of x and its derivatives')
plt.title("SELU Function and it's derivative")
plt.legend()
plt.show()
```
