```
from mxnet import nd
import matplotlib.pyplot as plt
```

## __Defining Inverse Square Root Linear Unit function__

```
def isru(x, alpha = 1.0):
    return x/(nd.sqrt(1 + alpha * nd.square(x)))
```

Below we are generating x values from range -10 to 10 with subsequent value of 0.1, that is last point -10, then -10 + 0.1 = -9.9, then -9.9 + 0.1 = -9.8.... and so on. This is what nd.arange function does

```
xs = nd.arange(-10, 10, 0.1)
ys = isru(xs)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(xs.asnumpy(), ys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of ISRU of x')
plt.title('ISRU Function')
plt.show()
```

## __Lets see how ISRU changes as we change alpha paramter__

```
alphas = nd.arange(0.1, 1.0, 0.1)
plt.figure(figsize = (10, 7))
for alpha in alphas:
    plt.plot(xs.asnumpy(), isru(xs, alpha).asnumpy(), label = f'Alpha: {alpha.asscalar():.1f}')
plt.legend()
plt.show()
```

So as we increases alpha the range output value shrinks within fixed intervals

## __Now defining derivative of isru funciton__

```
def isru_derivative(x, alpha = 1.0):
    return nd.square(1.0/nd.sqrt(1.0 + alpha * nd.square(x)))
```

```
dxs = nd.arange(-10, 10, 0.1)
dys = isru_derivative(xs)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(dxs.asnumpy(), dys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of derivatives of ISRU of x')
plt.title('ISRU Derivative')
plt.show()
```

## __Lets see how the derivative of ISRU changes as we change alpha paramter__

```
alphas = nd.arange(0.1, 1.0, 0.1)
plt.figure(figsize = (10, 7))
for alpha in alphas:
    plt.plot(xs.asnumpy(), isru_derivative(xs, alpha).asnumpy(), label = f'Alpha: {alpha.asscalar():.1f}')
plt.legend()
plt.show()
```

## __Plotting ISRU function and it's derivative together for alpha = 1.0__

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.figure(figsize = (10, 5))
plt.plot(xs.asnumpy(), ys.asnumpy(), label = 'ISRU function')
plt.plot(dxs.asnumpy(), dys.asnumpy(), label = 'ISRU function derivative')
plt.xlabel('Values of x')
plt.ylabel('Values of ISRU of x and its derivatives')
plt.title("ISRU Function and it's derivative")
plt.legend()
plt.show()
```
