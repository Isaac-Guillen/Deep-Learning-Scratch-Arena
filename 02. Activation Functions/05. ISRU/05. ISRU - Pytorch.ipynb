```
import torch
import matplotlib.pyplot as plt
```

## __Defining Inverse Square Root Linear Unit function__

```
def isru(x, alpha = 1.0):
    return x/(torch.sqrt(1 + alpha * torch.pow(x, 2)))
```

```
xs = torch.arange(-10, 10, 0.1)
ys = isru(xs)
```

```
plt.plot(xs, ys)
plt.xlabel('Values of x')
plt.ylabel('Values of ISRU of x')
plt.title('ISRU Function')
plt.show()
```

## __Lets see how ISRU changes as we change alpha paramter__

```
alphas = torch.arange(0.1, 1.0, 0.1)
plt.figure(figsize = (10, 7))
for alpha in alphas:
    plt.plot(xs, isru(xs, alpha), label = f'Alpha: {alpha:.1f}')
plt.legend()
plt.show()
```

So as we increases alpha the range output value shrinks within fixed intervals

## __Now defining derivative of isru funciton__

```
def isru_derivative(x, alpha = 1.0):
    return torch.pow(1.0/torch.sqrt(1.0 + alpha * torch.pow(x, 2)), 2)
```

```
dxs = torch.arange(-10, 10, 0.1)
dys = isru_derivative(xs)
```

```
plt.plot(dxs, dys)
plt.xlabel('Values of x')
plt.ylabel('Values of derivatives of ISRU of x')
plt.title('ISRU Derivative')
plt.show()
```

## __Lets see how the derivative of ISRU changes as we change alpha paramter__

```
alphas = torch.arange(0.1, 1.0, 0.1)
plt.figure(figsize = (10, 7))
for alpha in alphas:
    plt.plot(xs, isru_derivative(xs, alpha), label = f'Alpha: {alpha:.1f}')
plt.legend()
plt.show()
```

## __Plotting ISRU function and it's derivative together for alpha = 1.0__

```
plt.figure(figsize = (10, 5))
plt.plot(xs, ys, label = 'ISRU function')
plt.plot(dxs, dys, label = 'ISRU function derivative')
plt.xlabel('Values of x')
plt.ylabel('Values of ISRU of x and its derivatives')
plt.title("ISRU Function and it's derivative")
plt.legend()
plt.show()
```

Note the fact that maximum value of derivative is independent of the value of alpha
