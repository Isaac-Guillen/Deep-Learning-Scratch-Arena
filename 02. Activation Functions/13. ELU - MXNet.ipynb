```
from mxnet import nd
import matplotlib.pyplot as plt
```

## __Defining Exponential Linear Unit function__

```
def elu(x, alpha):
    return nd.where(x <= 0, alpha * (nd.exp(x) - 1.0), x)
```

```
alpha = 1.0
xs = nd.arange(-10, 10, 0.1)
ys = elu(xs, alpha)
print(f'alpha: {alpha}')
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(xs.asnumpy(), ys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of ELU of x')
plt.title('ELU Function')
plt.show()
```

## __Visiualizing elu with random alphas__

```
plt.figure(figsize = (10, 7))
alphas = nd.arange(0.1, 1.0, 0.1)
for alpha in alphas:
    plt.plot(xs.asnumpy(), elu(xs, alpha).asnumpy(), label = f'alpha: {alpha.asscalar():.2f}')
plt.legend()
plt.show()
```

## __Now defining derivative of elu funciton__

```
def elu_derivative(x, alpha):
    return nd.where(x <= 0, elu(x, alpha) + alpha, nd.ones_like(x))
```

```
dxs = nd.arange(-10, 10, 0.1)
dys = elu_derivative(dxs, alpha)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(dxs.asnumpy(), dys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of derivatives of ELU of x')
plt.title('ELU Derivative')
plt.show()
```

## __Plotting ELU function and it's derivative together__

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.figure(figsize = (10, 5))
plt.plot(xs.asnumpy(), ys.asnumpy(), label = 'ELU function')
plt.plot(dxs.asnumpy(), dys.asnumpy(), label = 'ELU function derivative')
plt.xlabel('Values of x')
plt.ylabel('Values of ELU of x and its derivatives')
plt.title("ELU Function and it's derivative")
plt.legend()
plt.show()
```
