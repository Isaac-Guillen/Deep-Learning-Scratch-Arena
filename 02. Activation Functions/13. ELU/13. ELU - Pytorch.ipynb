```
import torch
import matplotlib.pyplot as plt
```

## __Defining Exponential Linear Unit function__

```
def elu(x, alpha):
    return torch.where(x <= 0, alpha * (torch.exp(x) - 1.0), x)
```

```
alpha = 1.0
xs = torch.arange(-10, 10, 0.1)
ys = elu(xs, alpha)
print(f'alpha: {alpha}')
```

```
plt.plot(xs, ys)
plt.xlabel('Values of x')
plt.ylabel('Values of ELU of x')
plt.title('ELU Function')
plt.show()
```

## __Visiualizing elu with random alphas__

```
plt.figure(figsize = (10, 7))
alphas = torch.arange(0.1, 1.0, 0.1)
for alpha in alphas:
    plt.plot(xs, elu(xs, alpha), label = f'alpha: {alpha:.2f}')
plt.legend()
plt.show()
```

## __Now defining derivative of elu funciton__

```
def elu_derivative(x, alpha):
    return torch.where(x <= 0, elu(x, alpha) + alpha, torch.ones_like(x))
```

```
dxs = torch.arange(-10, 10, 0.1)
dys = elu_derivative(dxs, alpha)
```

```
plt.plot(dxs, dys)
plt.xlabel('Values of x')
plt.ylabel('Values of derivatives of ELU of x')
plt.title('ELU Derivative')
plt.show()
```

## __Plotting ELU function and it's derivative together__

```
plt.figure(figsize = (10, 5))
plt.plot(xs, ys, label = 'ELU function')
plt.plot(dxs, dys, label = 'ELU function derivative')
plt.xlabel('Values of x')
plt.ylabel('Values of ELU of x and its derivatives')
plt.title("ELU Function and it's derivative")
plt.legend()
plt.show()
```
