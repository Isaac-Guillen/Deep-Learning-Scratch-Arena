```
from mxnet import nd
import matplotlib.pyplot as plt
```

## __Defining Bent Identity function__

```
def bent_identity(x):
    return ((nd.sqrt(nd.square(x) + 1.0) - 1.0)/2.0) + x
```

```
xs = nd.arange(-10, 10, 0.1)
ys = bent_identity(xs)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(xs.asnumpy(), ys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of Bent Identity of x')
plt.title('Bent Identity Function')
plt.show()
```

## __Now defining derivative of bent identity funciton__

```
def bent_identity_derivative(x):
    return (x/(2 * nd.sqrt(nd.square(x) + 1))) + 1
```

```
dxs = nd.arange(-10, 10, 0.1)
dys = bent_identity_derivative(dxs)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(dxs.asnumpy(), dys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of derivatives of Bent Identity of x')
plt.title('Bent Identity Derivative')
plt.show()
```

## __Plotting Bent Identity function and it's derivative together__

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.figure(figsize = (10, 5))
plt.plot(xs.asnumpy(), ys.asnumpy(), label = 'Bent Identity function')
plt.plot(dxs.asnumpy(), dys.asnumpy(), label = 'Bent Identity function derivative')
plt.xlabel('Values of x')
plt.ylabel('Values of Bent Identity of x and its derivatives')
plt.title("Bent Identity Function and it's derivative")
plt.legend()
plt.show()
```
