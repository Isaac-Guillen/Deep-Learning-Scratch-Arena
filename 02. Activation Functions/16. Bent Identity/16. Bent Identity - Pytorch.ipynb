```
import torch
import matplotlib.pyplot as plt
```

## __Defining Bent Identity function__

```
def bent_identity(x):
    return ((torch.sqrt(torch.pow(x, 2) + 1.0) - 1.0)/2.0) + x
```

```
xs = torch.arange(-10, 10, 0.1)
ys = bent_identity(xs)
```

```
plt.plot(xs, ys)
plt.xlabel('Values of x')
plt.ylabel('Values of Bent Identity of x')
plt.title('Bent Identity Function')
plt.show()
```

## __Now defining derivative of bent identity funciton__

```
def bent_identity_derivative(x):
    return (x/(2 * torch.sqrt(torch.pow(x, 2) + 1))) + 1
```

```
dxs = torch.arange(-10, 10, 0.1)
dys = bent_identity_derivative(dxs)
```

```
plt.plot(dxs, dys)
plt.xlabel('Values of x')
plt.ylabel('Values of derivatives of Bent Identity of x')
plt.title('Bent Identity Derivative')
plt.show()
```

## __Plotting Bent Identity function and it's derivative together__

```
plt.figure(figsize = (10, 5))
plt.plot(xs, ys, label = 'Bent Identity function')
plt.plot(dxs, dys, label = 'Bent Identity function derivative')
plt.xlabel('Values of x')
plt.ylabel('Values of Bent Identity of x and its derivatives')
plt.title("Bent Identity Function and it's derivative")
plt.legend()
plt.show()
```
