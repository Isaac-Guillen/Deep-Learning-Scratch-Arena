```
from mxnet import nd
import matplotlib.pyplot as plt
```

## __Defining Soft Sign function__

```
def softsign(x):
    return x/(1.0 + nd.abs(x))
```

Below we are generating x values from range -10 to 10 with subsequent value of 0.1, that is last point -10, then -10 + 0.1 = -9.9, then -9.9 + 0.1 = -9.8.... and so on. This is what nd.arange function does

```
xs = nd.arange(-10, 10, 0.1)
ys = softsign(xs)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(xs.asnumpy(), ys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of Soft Sign of x')
plt.title('Soft Sign Function')
plt.show()
```

## __Now defining derivative of softsign funciton__

```
def softsign_derivative(x):
    return 1.0/nd.square(1.0 + nd.abs(x))
```

```
dxs = nd.arange(-10, 10, 0.1)
dys = softsign_derivative(dxs)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(dxs.asnumpy(), dys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of derivatives of Soft Sign of x')
plt.title('Soft Sign Derivative')
plt.show()
```

## __Plotting Soft Sign function and it's derivative together__

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.figure(figsize = (10, 5))
plt.plot(xs.asnumpy(), ys.asnumpy(), label = 'Soft Sign function')
plt.plot(dxs.asnumpy(), dys.asnumpy(), label = 'Soft Sign function derivative')
plt.xlabel('Values of x')
plt.ylabel('Values of Soft Sign of x and its derivatives')
plt.title("Soft Sign Function and it's derivative")
plt.legend()
plt.show()
```
