```
import torch
import matplotlib.pyplot as plt
```

## __Defining Rectified Linear Unit function__

```
def relu(x):
    # x[x < 0] = 0
    # return x
    # or
    return torch.max(x, torch.zeros_like(x))
```

```
xs = torch.arange(-10, 10, 0.1)
ys = relu(xs)
```

```
plt.plot(xs, ys)
plt.xlabel('Values of x')
plt.ylabel('Values of ReLU of x')
plt.title('ReLU Function')
plt.show()
```

## __Now defining derivative of relu funciton__

```
def relu_derivative(x):
    return torch.where(x > 0, torch.ones_like(x), torch.zeros_like(x))
```

```
dxs = torch.arange(-10, 10, 0.1)
dys = relu_derivative(dxs)
```

```
plt.plot(dxs, dys)
plt.xlabel('Values of x')
plt.ylabel('Values of derivatives of ReLU of x')
plt.title('ReLU Derivative')
plt.show()
```

## __Plotting ReLU function and it's derivative together__

```
plt.figure(figsize = (10, 5))
plt.plot(xs, ys, label = 'ReLU function')
plt.plot(dxs, dys, label = 'ReLU function derivative')
plt.xlabel('Values of x')
plt.ylabel('Values of ReLU of x and its derivatives')
plt.title("ReLU Function and it's derivative")
plt.legend()
plt.show()
```
