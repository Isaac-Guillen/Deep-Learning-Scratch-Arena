```
from mxnet import nd
import matplotlib.pyplot as plt
```

## __Defining ArcTan function__

Below we are generating x values from range -10 to 10 with subsequent value of 0.1, that is last point -10, then -10 + 0.1 = -9.9, then -9.9 + 0.1 = -9.8.... and so on. This is what nd.arange function does

```
xs = nd.arange(-10, 10, 0.1)
ys = nd.arctan(xs)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(xs.asnumpy(), ys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of ArcTan of x')
plt.title('ArcTan Function')
plt.show()
```

## __Now defining derivative of arctan funciton__

```
def arctan_derivative(x):
    return 1.0/((x**2) + 1.0)
```

```
dxs = nd.arange(-10, 10, 0.1)
dys = arctan_derivative(xs)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(dxs.asnumpy(), dys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of derivatives of ArcTan of x')
plt.title('ArcTan Derivative')
plt.show()
```

## __Plotting Arctan function and it's derivative together__

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.figure(figsize = (10, 5))
plt.plot(xs.asnumpy(), ys.asnumpy(), label = 'ArcTan function')
plt.plot(dxs.asnumpy(), dys.asnumpy(), label = 'ArcTan function derivative')
plt.xlabel('Values of x')
plt.ylabel('Values of ArcTan of x and its derivatives')
plt.title("ArcTan Function and it's derivative")
plt.legend()
plt.show()
```
