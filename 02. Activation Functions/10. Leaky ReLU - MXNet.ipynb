```
from mxnet import nd
import matplotlib.pyplot as plt
```

## __Defining Leaky Rectified Linear Unit function__

```
def leaky_relu(x, alpha = 0.2):
    return nd.maximum(x, alpha * x)
```

Below we are generating x values from range -10 to 10 with subsequent value of 0.1, that is last point -10, then -10 + 0.1 = -9.9, then -9.9 + 0.1 = -9.8.... and so on. This is what nd.arange function does

```
xs = nd.arange(-10, 10, 0.1)
ys = leaky_relu(xs)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(xs.asnumpy(), ys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of Leaky ReLU of x')
plt.title('Leaky ReLU Function')
plt.show()
```

## __Now defining derivative of leaky relu funciton__

```
def leaky_relu_derivative(x, alpha = 0.2):
    return nd.where(x > 0, nd.ones_like(x), nd.ones_like(x) * alpha)
```

```
dxs = nd.arange(-10, 10, 0.1)
dys = leaky_relu_derivative(dxs)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(dxs.asnumpy(), dys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of derivatives of Leaky ReLU of x')
plt.title('Leaky ReLU Derivative')
plt.show()
```

Note that mimum derivative is alpha(which is 0.2 in above case) not zero as in relu

## __Plotting Leaky ReLU function and it's derivative together__

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.figure(figsize = (10, 5))
plt.plot(xs.asnumpy(), ys.asnumpy(), label = 'Leaky ReLU function')
plt.plot(dxs.asnumpy(), dys.asnumpy(), label = 'Leaky ReLU function derivative')
plt.xlabel('Values of x')
plt.ylabel('Values of Leaky ReLU of x and its derivatives')
plt.title("Leaky ReLU Function and it's derivative")
plt.legend()
plt.show()
```
