```
import torch
import matplotlib.pyplot as plt
```

## __Defining Piecewise Linear Unit function__

```
def plu(x, alpha = 0.2, c = 1.0):
    return torch.max((alpha * (x + c)) - c, torch.min((alpha * (x - c)) + c, x))
```

```
xs = torch.arange(-10, 10, 0.1)
ys = plu(xs)
```

```
plt.plot(xs, ys)
plt.xlabel('Values of x')
plt.ylabel('Values of PLU of x')
plt.title('PLU Function')
plt.show()
```

## __Lets see how PLU changes when we take alpha constant and changes c and vice-versa__

```
print('Alpha is constant and equal to 0.2 and C changes')
cs = torch.arange(0.1, 1.0, 0.1)
plt.figure(figsize = (10, 7))
for c in cs:
    plt.plot(xs, plu(xs, alpha = 0.2, c = c), label = f'C: {c:.1f}')
plt.legend()
plt.show()
```

```
print('Now C is constant and equal to 1.0 and alpha changes')
alphas = torch.arange(0.1, 1.0, 0.1)
plt.figure(figsize = (10, 7))
for alpha in alphas:
    plt.plot(xs, plu(xs, alpha, c = 1.0), label = f'Alpha: {alpha:.1f}')
plt.legend()
plt.show()
```

## __Now defining derivative of plu funciton__

```
def plu_derivative(x, alpha = 0.2, c = 0.1):
    return torch.where(torch.abs(x) > c, torch.ones_like(x) * alpha, torch.ones_like(x))
```

```
dxs = torch.arange(-10, 10, 0.1)
dys = plu_derivative(dxs)
```

```
plt.plot(dxs, dys)
plt.xlabel('Values of x')
plt.ylabel('Values of derivatives of PLU of x')
plt.title('PLU Derivative')
plt.show()
```

## __Lets see how the derivative of PLU changes as we change alpha paramter__

```
alphas = torch.arange(0.1, 1.0, 0.1)
plt.figure(figsize = (10, 7))
for alpha in alphas:
    plt.plot(xs, plu_derivative(xs, alpha = alpha, c = 1.0), label = f'Alpha: {alpha:.1f}')
plt.legend()
plt.show()
```

```
cs = torch.arange(0.1, 1.0, 0.1)
plt.figure(figsize = (10, 7))
for c in cs:
    plt.plot(xs, plu_derivative(xs, alpha = 0.2, c = c), label = f'C: {c:.1f}')
plt.legend()
plt.show()
```

## __Plotting PLU function and it's derivative together for alpha = 0.2 and c = 1.0__

```
plt.figure(figsize = (10, 5))
plt.plot(xs, ys, label = 'PLU function')
plt.plot(dxs, dys, label = 'PLU function derivative')
plt.xlabel('Values of x')
plt.ylabel('Values of PLU of x and its derivatives')
plt.title("PLU Function and it's derivative")
plt.legend()
plt.show()
```
