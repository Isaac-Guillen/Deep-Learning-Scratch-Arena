```
from mxnet import nd
import matplotlib.pyplot as plt
```

## __Defining SoftPlus function__

```
def softplus(x):
    return nd.log(1.0 + nd.exp(x))
```

Below we are generating x values from range -10 to 10 with subsequent value of 0.1, that is last point -10, then -10 + 0.1 = -9.9, then -9.9 + 0.1 = -9.8.... and so on. This is what nd.arange function does

```
xs = nd.arange(-10, 10, 0.1)
ys = softplus(xs)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(xs.asnumpy(), ys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of SoftPlus of x')
plt.title('SoftPlus Function')
plt.show()
```

## __Now defining derivative of softplus funciton__

```
def softplus_derivative(x):
    return 1.0/(1.0 + nd.exp(-x))

# this derivative is nothing but a sigmoid function
```

```
dxs = nd.arange(-10, 10, 0.1)
dys = softplus_derivative(dxs)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(dxs.asnumpy(), dys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of derivatives of SoftPlus of x')
plt.title('SoftPlus Derivative')
plt.show()
```

## __Plotting SoftPlus function and it's derivative together__

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.figure(figsize = (10, 5))
plt.plot(xs.asnumpy(), ys.asnumpy(), label = 'SoftPlus function')
plt.plot(dxs.asnumpy(), dys.asnumpy(), label = 'SoftPlus function derivative')
plt.xlabel('Values of x')
plt.ylabel('Values of SoftPlus of x and its derivatives')
plt.title("SoftPlus Function and it's derivative")
plt.legend()
plt.show()
```
