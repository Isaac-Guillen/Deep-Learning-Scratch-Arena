```
import torch
import matplotlib.pyplot as plt
```

## __Defining SoftPlus function__

```
def softplus(x):
    return torch.log(1.0 + torch.exp(x))
```

```
xs = torch.arange(-10, 10, 0.1)
ys = softplus(xs)
```

```
plt.plot(xs, ys)
plt.xlabel('Values of x')
plt.ylabel('Values of SoftPlus of x')
plt.title('SoftPlus Function')
plt.show()
```

## __Now defining derivative of softplus funciton__

```
def softplus_derivative(x):
    return 1.0/(1.0 + torch.exp(-x))

# this derivative is nothing but a sigmoid function
```

```
dxs = torch.arange(-10, 10, 0.1)
dys = softplus_derivative(dxs)
```

```
plt.plot(dxs, dys)
plt.xlabel('Values of x')
plt.ylabel('Values of derivatives of SoftPlus of x')
plt.title('SoftPlus Derivative')
plt.show()
```

## __Plotting SoftPlus function and it's derivative together__

```
plt.figure(figsize = (10, 5))
plt.plot(xs, ys, label = 'SoftPlus function')
plt.plot(dxs, dys, label = 'SoftPlus function derivative')
plt.xlabel('Values of x')
plt.ylabel('Values of SoftPlus of x and its derivatives')
plt.title("SoftPlus Function and it's derivative")
plt.legend()
plt.show()
```
