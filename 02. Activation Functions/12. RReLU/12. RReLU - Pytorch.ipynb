```
import torch
import matplotlib.pyplot as plt
```

## __Defining Randomized Leaky Rectified Linear Unit function__

```
def rrelu(x, alpha):
    return torch.max(x, alpha * x)
```

```
alpha = 0.123
xs = torch.arange(-10, 10, 0.1)
ys = rrelu(xs, alpha)
print(f'alpha: {alpha}')
```

```
plt.plot(xs, ys)
plt.xlabel('Values of x')
plt.ylabel('Values of RReLU of x')
plt.title('RReLU Function')
plt.show()
```

## __Visiualizing rrelu with random alphas__

```
plt.figure(figsize = (10, 7))
alphas = torch.arange(0, 1.0, 0.1)
for alpha in alphas:
    plt.plot(rrelu(xs, alpha), label = f'alpha: {alpha:.2f}')
plt.legend()
plt.show()
```

## __Now defining derivative of rrelu funciton__

```
def rrelu_derivative(x, alpha):
    return torch.where(x > 0, torch.ones_like(x), torch.ones_like(x) * alpha)
```

```
dxs = torch.arange(-10, 10, 0.1)
dys = rrelu_derivative(dxs, alpha)
```

```
plt.plot(dxs, dys)
plt.xlabel('Values of x')
plt.ylabel('Values of derivatives of RReLU of x')
plt.title('RReLU Derivative')
plt.show()

print(f'Note that mimum derivative is alpha(which is {alpha:5f}... in above case) not zero as in relu')
```

## __Plotting RReLU function and it's derivative together__

```
plt.figure(figsize = (10, 5))
plt.plot(xs, ys, label = 'RReLU function')
plt.plot(dxs, dys, label = 'RReLU function derivative')
plt.xlabel('Values of x')
plt.ylabel('Values of RReLU of x and its derivatives')
plt.title("RReLU Function and it's derivative")
plt.legend()
plt.show()
```
