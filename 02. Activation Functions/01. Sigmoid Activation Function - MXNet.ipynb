```
from IPython.display import Image # this import is not important, its used for showing images in jupyter notebooks
from mxnet import nd
import matplotlib.pyplot as plt
```

## __Napier's Constant__

The value for napier's constant is 2.718281828

## __Defining sigmoid function__

```
def sigmoid(x):
    return 1.0/(1.0 + e**(-x))
```

Below we are generating x values from range -10 to 10 with subsequent value of 0.1, that is last point -10, then -10 + 0.1 = -9.9, then -9.9 + 0.1 = -9.8.... and so on. This is what nd.arange function does

```
global e # declaring value of napier's constan global so that it can be used inside the sigmoid function 
e = 2.718281828

xs = nd.arange(-10, 10, 0.1)
ys = sigmoid(xs)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(xs.asnumpy(), ys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of sigmoid of x')
plt.title('Sigmoid Function')
plt.show()
```

## __Now defining derivative of sigmoid funciton__

As we can see above that derivative of sigmoid of x is simply sigmoid of x multiplied by one minus sigmoid of x

```
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

```
dxs = nd.arange(-10, 10, 0.1)
dys = sigmoid_derivative(xs)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(dxs.asnumpy(), dys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of derivatives of sigmoid of x')
plt.title('Sigmoid Derivative')
plt.show()
```

## __Plotting sigmoid function and it's derivative together__

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.figure(figsize = (10, 5))
plt.plot(xs.asnumpy(), ys.asnumpy(), label = 'sigmoid function')
plt.plot(dxs.asnumpy(), dys.asnumpy(), label = 'sigmoid function derivative')
plt.xlabel('Values of x')
plt.ylabel('Values of sigmoid of x and its derivatives')
plt.title("Sigmoid Function and it's derivative")
plt.legend()
plt.show()
```

```
dys.max().asscalar() # getting the maximum value of the sigmoid derivatives
```
