```
from mxnet import nd
import matplotlib.pyplot as plt
```

## __Defining Swish function__

```
def sigmoid(x):
    return 1.0/(1.0 + nd.exp(-x))

def swish(x, beta = 1.0):
    return x * sigmoid(beta * x)
```

```
xs = nd.arange(-10, 10, 0.1)
ys = swish(xs)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(xs.asnumpy(), ys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of Swish of x')
plt.title('Swish Function')
plt.show()
```

## __Now defining derivative of swish funciton__

```
def swish_derivative(x):
    return swish(x) + (sigmoid(x) * (1 - swish(x)))
```

```
dxs = nd.arange(-10, 10, 0.1)
dys = swish_derivative(dxs)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(dxs.asnumpy(), dys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of derivatives of Swish of x')
plt.title('Swish Derivative')
plt.show()
```

## __Plotting Swish function and it's derivative together__

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.figure(figsize = (10, 5))
plt.plot(xs.asnumpy(), ys.asnumpy(), label = 'Swish function')
plt.plot(dxs.asnumpy(), dys.asnumpy(), label = 'Swish function derivative')
plt.xlabel('Values of x')
plt.ylabel('Values of Swish of x and its derivatives')
plt.title("Swish Function and it's derivative")
plt.legend()
plt.show()
```
