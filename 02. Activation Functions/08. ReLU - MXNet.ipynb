```
from mxnet import nd
import matplotlib.pyplot as plt
```

## __Defining Rectified Linear Unit function__

```
def relu(x):
    return nd.maximum(x, 0)
```

Below we are generating x values from range -10 to 10 with subsequent value of 0.1, that is last point -10, then -10 + 0.1 = -9.9, then -9.9 + 0.1 = -9.8.... and so on. This is what nd.arange function does

```
xs = nd.arange(-10, 10, 0.1)
ys = relu(xs)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(xs.asnumpy(), ys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of ReLU of x')
plt.title('ReLU Function')
plt.show()
```

## __Now defining derivative of relu funciton__

```
def relu_derivative(x, alpha = 0.2, c = 0.1):
    return nd.where(x > 0, nd.ones_like(x), nd.zeros_like(x))
```

```
dxs = nd.arange(-10, 10, 0.1)
dys = relu_derivative(dxs)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(dxs.asnumpy(), dys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of derivatives of ReLU of x')
plt.title('ReLU Derivative')
plt.show()
```

## __Plotting ReLU function and it's derivative together__

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.figure(figsize = (10, 5))
plt.plot(xs.asnumpy(), ys.asnumpy(), label = 'ReLU function')
plt.plot(dxs.asnumpy(), dys.asnumpy(), label = 'ReLU function derivative')
plt.xlabel('Values of x')
plt.ylabel('Values of ReLU of x and its derivatives')
plt.title("ReLU Function and it's derivative")
plt.legend()
plt.show()
```
