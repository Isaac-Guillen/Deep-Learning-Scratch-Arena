```
from mxnet import nd
import matplotlib.pyplot as plt
```

## __Defining Scaled Exponential Linear Unit function__

```
global lamda, alpha
lamda = 1.0507
alpha = 1.67326

def selu(x):
    return lamda * nd.where(x < 0, alpha * (nd.exp(x) - 1.0), x)
```

```
xs = nd.arange(-10, 10, 0.1)
ys = selu(xs)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(xs.asnumpy(), ys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of SELU of x')
plt.title('SELU Function')
plt.show()
```

## __Now defining derivative of selu funciton__

```
def selu_derivative(x):
    return lamda * nd.where(x < 0, alpha * nd.exp(x), nd.ones_like(x))
```

```
dxs = nd.arange(-10, 10, 0.1)
dys = selu_derivative(dxs)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(dxs.asnumpy(), dys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of derivatives of SELU of x')
plt.title('SELU Derivative')
plt.show()
```

## __Plotting SELU function and it's derivative together__

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.figure(figsize = (10, 5))
plt.plot(xs.asnumpy(), ys.asnumpy(), label = 'SELU function')
plt.plot(dxs.asnumpy(), dys.asnumpy(), label = 'SELU function derivative')
plt.xlabel('Values of x')
plt.ylabel('Values of SELU of x and its derivatives')
plt.title("SELU Function and it's derivative")
plt.legend()
plt.show()
```
