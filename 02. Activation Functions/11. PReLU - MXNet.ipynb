```
from mxnet import nd
import matplotlib.pyplot as plt
```

## __Defining Parametric Leaky Rectified Linear Unit function__

```
def prelu(x, alpha):
    return nd.maximum(x, alpha * x)
```

```
alpha = 0.2
xs = nd.arange(-1, 1, 0.1)
ys = prelu(xs, alpha)
print(f'alpha: {alpha}')
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.plot(xs.asnumpy(), ys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of PReLU of x')
plt.title('PReLU Function')
plt.show()
```

## __Visualizing prelu with different alphas__

```
plt.figure(figsize = (10, 7))
alphas = nd.arange(0.0, 1.0, 0.1)
for alpha in alphas:
    plt.plot(xs.asnumpy(), prelu(xs, alpha).asnumpy(), label = f'alpha: {alpha.asscalar():.2f}')
plt.legend()
plt.show()
```

## __Now defining derivative of prelu funciton__

```
def prelu_derivative(x, alpha):
    return nd.where(x > 0, nd.ones_like(x), nd.ones_like(x) * alpha)
```

```
dxs = nd.arange(-1, 1, 0.1)
dys = prelu_derivative(dxs, alpha)
```

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
alpha = 0.2
plt.plot(dxs.asnumpy(), dys.asnumpy())
plt.xlabel('Values of x')
plt.ylabel('Values of derivatives of PReLU of x')
plt.title('PReLU Derivative')
plt.show()

print(f'Note that mimum derivative is alpha(which is {alpha}... in above case) not zero as in relu')
```

## __Plotting PReLU function and it's derivative together__

```
# matplotlib doesn't support mxnet's ndarray but it supports numpy's ndarray, so we need to convert xs and ys to numpy ndarray by calling .asnumpy() method
plt.figure(figsize = (10, 5))
plt.plot(xs.asnumpy(), ys.asnumpy(), label = 'PReLU function')
plt.plot(dxs.asnumpy(), dys.asnumpy(), label = 'PReLU function derivative')
plt.xlabel('Values of x')
plt.ylabel('Values of PReLU of x and its derivatives')
plt.title("PReLU Function and it's derivative")
plt.legend()
plt.show()
```
