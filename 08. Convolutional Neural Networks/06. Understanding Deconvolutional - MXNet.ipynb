# MXNet (From Scratch)

```
import mxnet as mx
from mxnet import nd
import matplotlib.pyplot as plt

ctx = mx.gpu()
```

# Understanding Deconvolution

## Getting the data

```
mnist = mx.gluon.data.vision.MNIST()
```

```
image, _ = mnist[2]

image = image.flatten().astype('float32')

print(image.shape)

plt.imshow(image.asnumpy(), cmap = 'gray')
plt.show()
```

```
def deconvolution(image, kernel):
    filter_hight, filter_width = kernel.shape
    hight, width = (image.shape[0] + filter_hight) - 1, (image.shape[1] + filter_width) - 1
    layer = nd.zeros((int(hight), int(width)))

    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            layer[i: i + filter_hight, j: j + filter_width] += image[i, j] * kernel
    return layer
```

## Below we are generating image of shape (30, 30) from (28, 28) image, using a vertical kernel

```
kernal = nd.array([[1.0, 0.0, -1.0],
                   [1.0, 0.0, -1.0],
                   [1.0, 0.0, -1.0]])

output = deconvolution(image, kernal)

plt.imshow(output.asnumpy(), cmap = 'gray')
plt.show()

output.shape
```

## Below we are generating image of shape (30, 30) using a  horizontal kernel

```
kernal = nd.array([[1, 1, 1],
                   [0, 0, 0],
                   [-1, -1, -1]])

output = deconvolution(image, kernal)

plt.imshow(output.asnumpy(), cmap = 'gray')
plt.show()

output.shape # note the output shape
```

## Lets see some more examples with different types of filters.

```
kernal = nd.random.normal(0, 1, (3, 3))

output = deconvolution(image, kernal)

plt.imshow(output.asnumpy(), cmap = 'gray')
plt.show()

output.shape
```
